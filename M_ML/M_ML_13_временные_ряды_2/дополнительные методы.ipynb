```python
url = 'https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv'

df = pd.read_csv(url)

df.head()
```

```python
model = Prophet()
model.fit(df)
```

```python
future_df = model.make_future_dataframe(periods=365)

future_df.tail()
```
